```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.dbscan import *
from src.kmedoids import *
from src.kmeans import *
from sklearn.preprocessing import StandardScaler
```

```python
# dataset transforming and scaling
df = pd.read_csv('xclara.csv')

dataset = df.astype(float).values.tolist()

# scale it
dataset_std = StandardScaler().fit_transform(dataset)
```

```python
def visualize(points_colored):
    kategori = np.unique(points_colored)
    colors = ['#000000', '#A3523B', '#738A0B', '#29760A', '#31AF83', '#3181AF', '#5531AF', '#9B357C']

    for i in kategori:
        if(i < 0):
            color = '#000000'
        else:
            color = colors[i]

        x, y = [], []
        for j in range(len(dataset_std)):
            if a[j] == i:
                x.append(dataset_std[j, 0])
                y.append(dataset_std[j, 1])
        plt.scatter(x, y, s=20, c=color, alpha=1, marker='.')

    plt.show()
```

# dbscan gan (y)

```python
# fitting dbscan
a, b = fitDBScan(dataset_std, 0.2, 10)
visualize(a)
```
